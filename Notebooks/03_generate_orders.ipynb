```python
from pyspark.sql import SparkSession
from datetime import datetime, timedelta
import random

spark = SparkSession.builder.getOrCreate()

random.seed(200)

store_codes = ["CHN001","BLR001","HYD001","MUM001","DEL001"]
orders = []

for i in range(1,5001):
    order_date = (datetime(2025,1,1)+ timedelta(days=random.randint(0,180)))
    orders.append((i,random.randint(1,1000),random.randint(1,500),order_date.strftime("%Y-%m-%d"),
            random.randint(1,5),round(random.uniform(50,5000),2),random.choice(store_codes)))

df_orders = spark.createDataFrame(orders,["OrderID","CustomerID","ProductID","OrderDate","Quantity",
        "UnitPrice","StoreCode"])
df_orders.show(10)

print("Total Orders =",df_orders.count())
```

```python
orders_path = "/Volumes/retailanalytics/default/retail_volume/landing/orders"
(df_orders.coalesce(1).write.mode("overwrite").option("header", True).csv(orders_path))

print("Orders saved successfully")
```
